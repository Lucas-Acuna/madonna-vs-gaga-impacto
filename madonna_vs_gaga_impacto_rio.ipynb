```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv("madonna_gaga_impacto_rio.csv")

# Ver los primeros datos
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual para que quede más lindo
sns.set(style="whitegrid")

# Crear gráfico de barras
plt.figure(figsize=(8,5))
sns.barplot(x="Artista", y="Asistentes_estimados", data=df, palette="pastel")
plt.title("Cantidad de asistentes al show gratuito")
plt.ylabel("Asistentes estimados")
plt.xlabel("Artista")
plt.show()
```

```python
import pandas as pd

# Cargar el CSV
df = pd.read_csv("madonna_gaga_impacto_rio.csv")

# Verificamos que cargó bien
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.barplot(x="Artista", y="Asistentes_estimados", data=df, palette="pastel")
plt.title("Cantidad de asistentes al show gratuito")
plt.ylabel("Asistentes estimados")
plt.xlabel("Artista")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Artista", y="Ingreso_turismo_mill_usd", data=df, palette="pastel")
plt.title("Ingreso generado por turismo en Río de Janeiro (millones USD)")
plt.ylabel("Ingreso turístico (en millones de USD)")
plt.xlabel("Artista")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Artista", y="Gasto_prom_persona_usd", data=df, palette="pastel")
plt.title("Gasto promedio por persona durante el evento (USD)")
plt.ylabel("Gasto por persona (USD)")
plt.xlabel("Artista")
plt.show()
```

```python
plt.figure(figsize=(10,4))

# Gráfico 1: Cobertura en medios
plt.subplot(1, 2, 1)
sns.barplot(x="Artista", y="Cobertura_medios", data=df, estimator=lambda x: len(x), ci=None, palette="pastel")
plt.title("Nivel de cobertura mediática")
plt.ylabel("Nivel")
plt.xlabel("Artista")

# Gráfico 2: Impacto global
plt.subplot(1, 2, 2)
sns.barplot(x="Artista", y="Impacto_Global", data=df, estimator=lambda x: len(x), ci=None, palette="pastel")
plt.title("Nivel de impacto global")
plt.ylabel("Nivel")
plt.xlabel("Artista")

plt.tight_layout()
plt.show()
```
